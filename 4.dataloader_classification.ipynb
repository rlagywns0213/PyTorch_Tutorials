```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
# 뉴럴 네트워크에 입력값으로 array 형태로 변환
data['data'] = data['data'].astype(np.float32)
data['target'] = data['target'].astype(np.float32)

x_data = from_numpy(data['data'])
y_data = from_numpy(data['target'])
y_data = y_data.reshape(-1,1)
print(x_data.shape, y_data.shape)
```

```python
class Model(nn.Module):
    def __init__(self): #생성자 (필요한 layer 정의)
        super(Model, self).__init__() #항상 들어가는 부분으로 base에 생성자 호출
        self.l1 = nn.Linear(30,256) #입력 노드  맞게끔 변환
        self.l2 = nn.Linear(256,256)
        self.l3 = nn.Linear(256,128)
        self.l4 = nn.Linear(128,128)
        self.l5 = nn.Linear(128,64)
        self.l6 = nn.Linear(64,32)
        self.l7 = nn.Linear(32,16)
        self.l8 = nn.Linear(16,8)
        self.l9 = nn.Linear(8,6)
        self.l10 = nn.Linear(6,6)
        self.l11 = nn.Linear(6,4)
        self.l12 = nn.Linear(4,1)
        self.ReLU = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

# 오버라이딩 : 부모클래스에 있는 함수를 자식 클래스에서 재정의해서 사용 (특화된 기능 하도록)
    def forward(self, x): 
        out1 = self.ReLU(self.l1(x))
        out2 = self.ReLU(self.l2(out1))
        out3 = self.ReLU(self.l3(out2))
        out4 = self.ReLU(self.l4(out3))
        out5 = self.ReLU(self.l5(out4))
        out6 = self.ReLU(self.l6(out5))
        out7 = self.ReLU(self.l7(out6))
        out8 = self.ReLU(self.l8(out7))
        out9 = self.ReLU(self.l9(out8))
        out10 = self.ReLU(self.l10(out9))
        out11 = self.ReLU(self.l11(out10))
        
        y_pred = self.sigmoid(self.l12(out11))
        return y_pred

model = Model()        

#loss 정의
criterion = nn.BCELoss(reduction ='mean')
optimizer = optim.Adagrad(model.parameters(), lr=0.01)

for epoch in range(2000):
    y_pred = model(x_data)
    #loss 계산 후 출력
    loss = criterion(y_pred, y_data)
    print(f'Epoch: {epoch+1}/2000 | Loss: {loss.item():.4f}')
    # gradient 값이 계속 누적되므로 0으로 reset해준다. (autograd 위함)
    optimizer.zero_grad()
    loss.backward() #학습에 필요한 모든 gradient 한 번에 구해지고 
    optimizer.step() # step을 통해 adagrad 방법으로 학습이 된다.( 학습률 0.01 )
```

```python
df = pd.read_csv('/content/drive/MyDrive/4-1 실전기계학습/수업자료/data/Titanic_train.csv')
df.drop(['PassengerId','Name','Ticket','Cabin'], axis=1,inplace=True) #기본 인덱스 삭제
x_data = df.drop('Survived', axis=1)
y_data = df['Survived']
```

```python
x_data.isnull().sum()
x_data.fillna(0, inplace=True)
```

```python
# object column은 onehotencoding 실시
object_cols = list(x_data.select_dtypes('object').columns)
x_data_onehot = pd.get_dummies(x_data[object_cols], drop_first=False)

x_data.drop(object_cols, axis=1, inplace=True)
x_data = pd.concat([x_data, x_data_onehot], axis=1)

print(object_cols)
x_data
```

```python
from torch import nn, optim, from_numpy, tensor
import numpy as np
import pandas as pd
from torch.utils.data import Dataset, DataLoader 
import torch

class trainData(Dataset):
    
    def __init__(self, x_data, y_data):
        self.x_data = x_data
        self.y_data = y_data
        
    def __getitem__(self, index):
        return self.x_data[index], self.y_data[index]
        
    def __len__ (self):
        return len(self.x_data)


train_data = trainData(torch.FloatTensor(x_data.values), 
                        torch.FloatTensor(y_data.values).reshape(-1,1))

train_loader = DataLoader(dataset=train_data, 
                          batch_size=32, 
                          shuffle=True,
                          num_workers=2)
```

```python
class Model(nn.Module):
    def __init__(self): #생성자 (필요한 layer 정의)
        super(Model, self).__init__()
        self.l1 = nn.Linear(11,64)
        self.l2 = nn.Linear(64,128)
        self.l3 = nn.Linear(128,64)        
        self.l4 = nn.Linear(64,32)
        self.l5 = nn.Linear(32,1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x): 
        out1 = self.sigmoid(self.l1(x))
        out2 = self.sigmoid(self.l2(out1))
        out3 = self.sigmoid(self.l3(out2))
        out4 = self.sigmoid(self.l4(out3))
        y_pred = self.sigmoid(self.l5(out4))
        return y_pred

model = Model()        

#loss 정의
criterion = nn.BCELoss(reduction ='mean')
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
for epoch in range(100):
    for i, data in enumerate(train_loader, 0): #0번째 부터 시작한다.
        inputs, labels = data
        y_pred = model(inputs)
        loss = criterion(y_pred, labels)
        print(f'Epoch: {epoch+1} | Batch {i+1} | Loss {loss.item():.4f}')
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```

```python
from torch import nn, optim, from_numpy, tensor
import numpy as np
import pandas as pd
from torch.utils.data import Dataset, DataLoader 
import torch

df = pd.read_csv('/content/drive/MyDrive/4-1 실전기계학습/수업자료/data/otto_train.csv')
df.drop('id',axis=1,inplace=True)
df['target'] =  df['target'].map({'Class_1': 1, 'Class_2': 2,
                                          'Class_3': 3, 'Class_4': 4,
                                          'Class_5': 5, 'Class_6': 6,
                                          'Class_7': 7, 'Class_8': 8,
                                          'Class_9': 9})
df['target'] = df['target'].astype('float64')
x_data = df.drop('target', axis=1)
x_data
```

```python
class OttoDataset(Dataset):
    def __init__(self):
        self.len = len(df)
        self.x_data =from_numpy(x_data.values)
        self.y_data = from_numpy(df['target'].values)

    def __getitem__(self, index):
        return self.x_data[index], self.y_data[index]

    def __len__(self):
        return self.len

dataset = OttoDataset()

train_loader = DataLoader(dataset=dataset,
                          batch_size=128,
                          shuffle=True,
                          num_workers=2)
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        self.l1 = nn.Linear(93, 128)
        self.l2 = nn.Linear(128, 128)
        self.l3 = nn.Linear(128, 64)
        self.l4 = nn.Linear(64,32)
        self.l5 = nn.Linear(32,9)
    def forward(self,x):
        out1 = F.relu(self.l1(x))
        out2 = F.relu(self.l2(out1))
        out3 = F.relu(self.l3(out2))
        out4 = F.relu(self.l4(out3))
        return self.l5(out4)

model = Model()

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.5)
```

```python
from torch.autograd import Variable
import torch.nn.functional as F
for epoch in range(10):
    model.train()
    for i, (data,target) in enumerate(train_loader):
        data, target = Variable(data).float(),Variable(target).type(torch.LongTensor)
        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output, target-1)
        print(f'Epoch: {epoch+1} | Batch {i+1} | Loss {loss.item():.4f}')
        loss.backward()
        optimizer.step()
```
